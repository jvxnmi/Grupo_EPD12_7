```python
import pandas as pd
```

```python
paises = 'https://raw.githubusercontent.com/jvxnmi/fFSI/refs/heads/main/EPD12_7_happyscore_income.csv'
df1 = pd.read_csv(paises)
df1
```

Realice una operaciÃ³n donde se seleccionen varias columnas relevantes, pero mostrando
solo algunas de las filas, es decir realizando un filtrado en los valores de al menos dos
columnas que no tienen por quÃ© coincidir con las que se muestren.


Mostramos paises con hayppyScore descendente filtrando por ingreso medio entre 1000 y 20500

```python
df1["avg_income"] = pd.to_numeric(df1["avg_income"], errors='coerce')

filtered_df = df1[(df1["avg_income"] >= 100) & (df1["avg_income"] <= 20500)]
sorted_df = filtered_df.sort_values(by="happyScore", ascending=False)
resultado = sorted_df["country"]
resultado
```

Construya y muestre un nuevo dataframe que agregue al original una nueva columna
calculada numÃ©rica a partir de una o varias de las ya existentes. Posteriormente agrupar

Creamos un dataframe nuevo df2 copiando el original (df1) y creamos una nueva columna llamada Realcion satisfaccion/ingresos que divide  avg_satisfaction con avg_income

```python
df2 = df1[(df1['avg_income'] != 0)]
```

```python
df2['RelaciÃ³n SatisfacciÃ³n/Ingresos'] = df2['avg_satisfaction'] / df2['avg_income'] 
df2[['country','avg_satisfaction','avg_income','RelaciÃ³n SatisfacciÃ³n/Ingresos']]
```

Creamos otro daataframe df3 que une df1 y df2

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df3.groupby('region')[['country', 'avg_satisfaction', 'RelaciÃ³n SatisfacciÃ³n/Ingresos', 'happyScore']].sum().sort_values('happyScore', ascending=False)
```

Finalmente agrupamos por regiones y ordenamos por hayppyScore descendente
